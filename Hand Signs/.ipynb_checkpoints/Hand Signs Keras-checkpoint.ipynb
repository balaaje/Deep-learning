```python
import pandas as pd
import numpy as np
import h5py

#Tensorflow
import tensorflow as tf
from tensorflow import keras
from keras.utils import to_categorical

#importing the required keras modules containing models and layers
from keras.models import Sequential
from keras.layers import Dense,Conv2D,Dropout,Flatten,MaxPooling2D
from keras.preprocessing import image
from keras.applications.imagenet_utils import preprocess_input

import  matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
print(tf.__version__)
```

# Importing Hand signs DataSet

```python
def load_dataset():
    train_dataset = h5py.File('train_signs.h5', "r")
    train_set_x_orig = np.array(train_dataset["train_set_x"][:]) # your train set features
    train_set_y_orig = np.array(train_dataset["train_set_y"][:]) # your train set labels

    test_dataset = h5py.File('test_signs.h5', "r")
    test_set_x_orig = np.array(test_dataset["test_set_x"][:]) # your test set features
    test_set_y_orig = np.array(test_dataset["test_set_y"][:]) # your test set labels

    classes = np.array(test_dataset["list_classes"][:]) # the list of classes
    
    train_set_y_orig = train_set_y_orig.reshape((1, train_set_y_orig.shape[0]))
    test_set_y_orig = test_set_y_orig.reshape((1, test_set_y_orig.shape[0]))
    
    return train_set_x_orig, train_set_y_orig, test_set_x_orig, test_set_y_orig, classes
```

```python
# Loading the data (signs)
X_train_orig, Y_train_orig, X_test_orig, Y_test_orig, classes = load_dataset()
```

```python
train_images=train_images_orig/255
test_images=test_images_orig/255
```

```python
input_shape=train_images.shape[1:]
print(input_shape)
```

```python
train_labels = to_categorical(Y_train_orig.reshape(-1))
#print(train_labels)
print(train_labels.shape)
```

```python
test_labels = to_categorical(Y_test_orig.reshape(-1))
#print(test_labels)
print(test_labels.shape)
```

```python
train_images.shape
```

```python
model=Sequential()
model.add(Conv2D(filters=50, kernel_size=(3,3), input_shape=input_shape,padding="same"))
model.add(Dense(128, activation=tf.nn.relu))
model.add(MaxPooling2D(pool_size=(8, 8),strides=(8,8),padding="same"))
model.add(Conv2D(filters=50, kernel_size=(3,3),padding="same"))
model.add(Dense(128, activation=tf.nn.relu))
model.add(MaxPooling2D(pool_size=(4, 4),strides=(4,4),padding="same"))
model.add(Flatten()) # Flattening the 2D arrays for fully connected neural layers.
model.add(Dropout(0.2))
model.add(Dense(6,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', 
              loss='categorical_crossentropy', 
              metrics=['accuracy'])
```

```python
model.fit(x=train_images,y=train_labels, epochs=10)
```

```python
preds = model.evaluate(test_images, test_labels)
print ("Loss = " + str(preds[0]))
print ("Test Accuracy = " + str(preds[1]))
```

```python
from keras import applications
from keras.models import Sequential,Model,load_model
from keras import optimizers
from keras.layers import GlobalMaxPooling2D,GlobalAveragePooling2D
```

```python
img_height,img_width = 64,64 
num_classes = 6
#If imagenet weights are being loaded, 
#input must have a static square shape (one of (128, 128), (160, 160), (192, 192), or (224, 224))
base_model = applications.resnet50.ResNet50(weights= None, input_shape= (img_height,img_width,3))
```

```python
x = base_model.output
predictions = Dense(num_classes, activation= 'softmax')(x)
model = Model(inputs = base_model.input, outputs = predictions)
```

```python
adam = optimizers.Adam(lr=0.0001)
```

```python
model.compile(optimizer=adam, 
              loss='categorical_crossentropy', 
              metrics=['accuracy'])
```
